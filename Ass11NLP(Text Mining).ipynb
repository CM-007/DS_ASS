```python
import pandas as pd
import numpy as np
import string
from matplotlib.pyplot import imread
from matplotlib import pyplot as plt
```

```python
musk_data=pd.read_csv("Elon_musk.csv",encoding="latin1",error_bad_lines=False)
```

```python
musk_data
```

```python
musk_data.drop(["Unnamed: 0"],inplace=True,axis=1)
```

```python
musk_data
```

```python
musk_data=[Text.strip() for Text in musk_data.Text] 
musk_data=[Text for Text in musk_data if Text] 
musk_data[0:10]
```

```python
text = ' '.join(musk_data)
text
```

```python
no_punc_text = text.translate(str.maketrans('', '', string.punctuation))
no_punc_text
```

```python
import re
no_url_text=re.sub(r'http\S+', '', no_punc_text)
no_url_text
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
text_tokens = word_tokenize(no_url_text)
print(text_tokens[0:50])
```

```python
len(text_tokens)
```

```python
my_stop_words = stopwords.words('english')
my_stop_words.append('the')
no_stop_tokens = [word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens[0:60])
```

```python
lower_words = [Text.lower() for Text in no_stop_tokens]
print(lower_words[0:100])
```
